# Importing the Library

```python
from bs4 import BeautifulSoup as bs
```

```python
import requests
```

### Goal :-   The main goal of this project is to Ensure & collect high-quality data that can be used to make informed decisions and improve an organizationâ€™s operational processes .

```python
url="https://www.flipkart.com/hp-15s-ryzen-3-dual-core-3250u-4-gb-1-tb-hdd-windows-10-home-15s-gr0006au-thin-light-laptop/product-reviews/itmdbb2d234797b6?pid=COMFZHZFZFFBJBHH&lid=LSTCOMFZHZFZFFBJBHH7WNMPE&marketplace=FLIPKART"
```

```python
r=requests.get(url)
r
```

```python
htmlcontent=r.content
htmlcontent
```

```python
soup=bs(htmlcontent,"html.parser")
```

```python
print(soup.prettify())
```

# Title Of the Product

```python
title=soup.title
print(title)
print(type(title))
```

```python
print(title.string)
```

# Check the Product Price

```python
price=soup.find_all("div",class_="_30jeq3")
```

```python
price
```

```python
product_price=[]

for i in range(0,len(price)):
    product_price.append(price[i].get_text())
```

```python
product_price
```

# Get all the customer names

```python
names=soup.find_all("p",class_="_2sc7ZR _2V5EHH")
names
```

```python
cust_name=[]

for i in range(0,len(names)):
    cust_name.append(names[i].get_text())
```

```python
cust_name
```

# Get All the Customers Comments

```python
comment=soup.find_all("p",class_="_2-N8zT")
comment
```

```python
cust_comment=[]

for i in range(0,len(comment)):
    cust_comment.append(comment[i].get_text())
```

```python
cust_comment
```

# Check how much star gives customer to this product

```python
star=soup.find_all("div",class_="_3LWZlK _1BLPMq")
star
```

```python
cust_star=[]

for i in range(0,len(star)):
    cust_star.append(star[i].get_text())
```

```python
cust_star
```

# Get all the customers review to this product

```python
review=soup.find_all("div",class_="t-ZTKy")
review
```

```python
cust_review=[]

for i in range(0,len(review)):
    cust_review.append(review[i].get_text())
```

```python
cust_review
```

# Applying Tabular Data Structure

```python
import pandas as pd
```

```python
df=pd.DataFrame()
df["Customer_Names"]=cust_name
df["Comments"]=cust_comment
df["Stars"]=cust_star
df["Reviews"]=cust_review
```

```python
df
```

# Exporting the dataset into csv file

```python
df.to_csv("Hp Filpkart Product Review",index=False)
```
